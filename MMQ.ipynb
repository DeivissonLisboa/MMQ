# **Método dos mínimos quadrados**

```python
# Importar bibliotecas para uso de algoritmos matemáticos
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
# Definindo lista de pontos onde cada elemento da lista
# é um vetor de coordenadas x e y
pontos = np.array([
  # [x, y]
  [1, 1],
  [3, 2],
  [4, 4],
  [6, 4],
  [8, 5],
  [9, 7],
  [11, 8],
  [14, 9]
])
```

## Gerando gráfico para observar se há um comportamento linear

```python
# Definindo a função gráfico
# O único argumento extritamente necessário é a lista
# lista de pontos.
def grafico(
  pontos, 
  titulo_grafico="Gráfico Linear", 
  titulo_x="Medidas X", 
  titulo_y="Medidas Y",
  cor="black",
  linhas=False
):
  x_i = pontos[:, 0]
  y_i = pontos[:, 1]

  fig = plt.figure()
  ax = fig.add_axes([0, 0, 1, 1])
  ax.scatter(x = x_i, y = y_i, color = cor)
  plt.title(titulo_grafico)
  plt.xlabel(titulo_x)
  plt.ylabel(titulo_y)
  
  if linhas:
    plt.plot(x_i, y_i, color=cor)

  plt.show()
```

```python
# Chamando a função gráfico.
# Os argumentos devem ser colocados na ordem que 
# foram definidos acima!
grafico(pontos)
```

```python
# Gráfico conectando linhas e pontos
grafico(pontos, linhas=True)
```

## Entendendo um comportamento próximo do linear aplica-se o MMQ
