```python
# Importar bibliotecas para uso de algoritmos matemáticos
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo lista de pontos onde cada elemento da lista
# é um vetor de coordenadas x e y
pontos = np.array([
  # [x, y]
  [1, 1],
  [3, 2],
  [4, 4],
  [6, 4],
  [8, 5],
  [9, 7],
  [11, 8],
  [14, 9]
])
```

**Gerando gráfico para observar se há um comportamento linear

```python
# Definindo a função gráfico
# O único argumento extritamente necessário é a lista
# lista de pontos.
def grafico(
  pontos, 
  titulo_grafico="Gráfico Linear", 
  titulo_x="Medidas X", 
  titulo_y="Medidas Y",
  cor_pontos="black"
):
  fig = plt.figure()
  ax = fig.add_axes([0, 0, 1, 1])
  ax.scatter(x = pontos[:, 0], y = pontos[:, 1], color = cor_pontos)
  plt.title(titulo_grafico)
  plt.xlabel(titulo_x)
  plt.ylabel(titulo_y)
  plt.show()
```

```python
# Chamando a função gráfico.
# Os argumentos devem ser colocados na ordem que 
# foram definidos acima!
grafico(pontos)
```

```python
# Entendendo um comportamento próximo do linear aplica-se o MMQ

#n significa o número de medida das variáveis
n=8

#Obtenção dos termos Xi^2 e Xi*Yi
Xi2=Xi**2
XiYi=Xi*Yi

#Cálculo dos coeficientes a e b pelo MMQ
a = (sum(Xi)*sum(Yi) -n*sum(XiYi))/(sum(Xi)**2 -n*sum(Xi2))
b = (sum(XiYi)*sum(Xi) - sum(Xi2)*sum(Yi))/(sum(Xi)**2 -n*sum(Xi2))
print('O somatorio de X:',sum(Xi),"\n",'O somatorio de Y:',sum(Yi),"\n",'O somatorio de X*Y:',sum(XiYi),"\n",'O somatorio de X^2:',sum(Xi2),"\n")
print(f'O valor da constante a é: {a:,.2f}',"\n",f'O valor da constante b é: {b:,.2f}')
```

```python
# Criar lista de valores ajustados
Ya=[]
for i in range(n):
  y= a*Xi[i] + b
  Ya.append(y)
```
